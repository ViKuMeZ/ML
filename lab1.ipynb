```python
sns.histplot(df['hum'], kde=True)
plt.title('Распределение hum')
plt.xlabel('hum')
plt.ylabel('Частота')
plt.show()
```

```python
sns.countplot(x='windspeed', data=df)
plt.title('Распределение windspeed')
plt.xlabel('windspeed')
plt.ylabel('Количество')
plt.show()
```

```python
sns.boxplot(x=df['temp'])
plt.title('Boxplot temp')
plt.xlabel('temp')
plt.show()
```

Распределение hum не соответсвует нормальному распределению, как и распределение windspeed, а вот temp выглядит как правда.
